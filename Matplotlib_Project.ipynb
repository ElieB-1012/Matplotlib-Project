# **Project 1: Plotting Einstein's Equation**

```python
import numpy as np
```

$$E=mc^2$$


With NumPy I created two arrays: E and m, where **m** is simply 11 evenly spaced values representing 0 to 10 grams. E must be the energy equivalent to mass.

```python
m = np.linspace(0,10,11)
c = 3 * 10**8
E = m * c**2
```

```python
E
```

```python
import matplotlib.pyplot as plt
import math
```

## **Functionality Version**

```python
plt.plot(m,E, color = 'red', lw = 5);
plt.title('E = mc^2', color = 'red')
plt.xlabel('masse (g)')
plt.ylabel('Energy (J)')
plt.xlim(0,10);
```

## **OOP Version**

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E, color = 'red', lw = 5)
axes.set_title('E = mc^2', color = 'red')
axes.set_xlabel('masse (g)')
axes.set_ylabel('Energy (J)')
axes.set_xlim(0,10);
```

## **Logarithm Version**

```python
plt.plot(m,E, color = 'red', lw = 5);
plt.title('E = mc^2', color = 'red')
plt.xlabel('masse (g)')
plt.ylabel('Energy (J)')
plt.xlim(0,10);
plt.yscale('log')
plt.grid(which = 'both', axis = 'y')
```

# **Project 2: Creating plots from data points**

In finance, the yield curve is a curve showing multiple yields to maturity or interest rates over different contract durations (2 months, 2 years, 20 years, etc.) for a similar loan contract. The curve shows the relationship between the (level of) interest rate (or cost of borrowing) and the time to maturity, known as the "term", of a given borrower's debt in a given currency .

The US dollar interest rates paid on US Treasury securities for various maturities are closely watched by many traders and are usually plotted on a graph such as the one to the right, informally called "the yield curve".

The yield curve (Yield Curve) is a graphical representation of the difference between the interest rates of short-term and long-term government bonds, called "treasury bills". Indeed, all countries have a yield curve for their bond markets.


### Part 1: Yield curve data

We obtained yield curve data for you from the [US Department of Treasury](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx? data=yield). The data shows the interest paid for a US Treasury bond for a certain contract term. The label list displays the corresponding contract length by index position.

Given Data:

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

### Part 2: Drawing Graphs

### With Figure:

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007,label='july16_2007')
axes.plot(labels,july16_2020,label='july16_2020')
plt.legend()
plt.show()
```

### With Subplot:

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(12,8))

axes[0].plot(labels, july16_2007,label='july16_2007')
axes[0].set_title("16 juillet 2007")
axes[1].plot(labels,july16_2020,label='july16_2020')
axes[1].set_title("16 juillet 2020")


plt.show()
```
